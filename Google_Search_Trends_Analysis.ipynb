## Importing Required Libraries

```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
from wordcloud import WordCloud
trends = TrendReq()
```

## Let's Check what is Today's trending search in India

```python
india = trends.trending_searches(pn='india')
india=str(india)
wordcloud = WordCloud(collocations = False, background_color = 'white').generate(india)
plt.figure(figsize = (10, 10), facecolor = None)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

## Let's check Corona Search By Region

```python
trends.build_payload(kw_list=["coronavirus"])
data = trends.interest_by_region()
data = data.sort_values(by="coronavirus", ascending=False)
data = data.head(20)
print(data)
```

```python
data.reset_index().plot(x="geoName", y="coronavirus", figsize=(20,15), kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```

## Let's check Trending Search in some of technologies over years

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=["Machine Learning","Artificial Intelligence","Data Analytics"])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(20, 15))
data["Data Analytics"].plot(label="Data Analytics"),data["Machine Learning"].plot(label="Machine Learning"),data["Artificial Intelligence"].plot(label="Artificial Intelligence")
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for Data Analytics', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.legend()
plt.show()
```
